```python
import pandas as pd

# Load each data file into a DataFrame
data_file1 = pd.read_csv("C:\\Users\\admin\\Downloads\\data_1.csv\\data_1.csv")
data_file2 = pd.read_csv("C:\\Users\\admin\\Downloads\\data_2.csv\\data_2.csv")
data_file3 = pd.read_csv("C:\\Users\\admin\\Desktop\\python workspace\\finaldataset_3.csv")
data_file4 = pd.read_csv("C:\\Users\\admin\\Downloads\\data_4.csv\\data_4.csv")
```

```python
# Concatenate the dataframes
merged_data = pd.concat([data_file1, data_file2, data_file3, data_file4], ignore_index=True)
```

```python
merged_data.drop('YearOfRelease', axis=1, inplace=True)
```

```python
# Assuming 'merged data' is your DataFrame and 'Customer_Id' is the column containing user IDs

import pandas as pd

# Define aggregation functions
f = ['count', 'mean']

# Group by 'Customer_Id' and aggregate ratings
dataset_cust_summary = merged_data.groupby('CustomerID')['Rating'].agg(f)

# Convert index to integer
dataset_cust_summary.index = dataset_cust_summary.index.map(int)

# Calculate the threshold for active users (quantile 0.8)
cust_benchmark = round(dataset_cust_summary['count'].quantile(0.8), 0)

# Extract inactive users (users who rate less often)
inactive_users = dataset_cust_summary[dataset_cust_summary['count'] < cust_benchmark].index

print(f'Customer minimum times of review: {cust_benchmark}')

# Create a list of active users (users who rate more often)
active_users = dataset_cust_summary[dataset_cust_summary['count'] >= cust_benchmark].index

# Print the list of active users
print("Active Users (Users who rate more often):")
print(active_users)

# Print the list of inactive users
print("Inactive Users (Users who rate less often):")
print(inactive_users)

num_active_users = len(active_users)
num_inactive_users = len(inactive_users)

print(f'Number of active users: {num_active_users}')
print(f'Number of inactive users: {num_inactive_users}')
```

```python
def recommend(movie_title, rating=5):
    if movie_title in similar_movie_df.index:
        score = similar_movie_df[movie_title]*(rating-2.5)
        score = score.sort_values(ascending = False)[1:6]    #recommend top 5 movies

        return score

    else:
        top_movies = unique_ratings[:6]
        print(f'"{movie_title}" not found in the database')
        print("Watch these instead:")
        return top_movies[['Title', 'num_ratings']]
```
